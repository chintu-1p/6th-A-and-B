```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler, MaxAbsScaler, RobustScaler

input_file = '/kaggle/input/10000-rows-randomly-sampled-from-lcallcsv/lc_10000_rows.csv'
dataset = pd.read_csv(input_file)

numerical_columns = dataset.select_dtypes(include=['float64', 'int64']).dropna(axis=1)

scalers = {
    'MinMaxScaler': MinMaxScaler(),
    'ZScoreScaler': StandardScaler(),
    'MaxAbsScaler': MaxAbsScaler(),
    'RobustScaler': RobustScaler()
}

output_path = '/kaggle/working'
for scaler_name, scaler in scalers.items():
    scaled_data = scaler.fit_transform(numerical_columns)
    scaled_df = pd.DataFrame(scaled_data, columns=numerical_columns.columns)
    scaled_df.to_csv(f'{output_path}/{scaler_name.lower()}_scaled.csv', index=False)
```
