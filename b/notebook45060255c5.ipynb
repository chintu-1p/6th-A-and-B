```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

input_file = '/kaggle/input/100-sales/100_Sales.csv'
dataset = pd.read_csv(input_file)

numerical_columns = dataset.select_dtypes(include=['float64', 'int64'])

scaler = StandardScaler()
zscore_scaled_data = scaler.fit_transform(numerical_columns)
zscore_scaled_df = pd.DataFrame(zscore_scaled_data, columns=numerical_columns.columns)

mean_centered_data = numerical_columns - numerical_columns.mean()
mean_centered_df = mean_centered_data

zscore_scaled_df.to_csv('zscore_scaled_sales.csv', index=False)
mean_centered_df.to_csv('mean_centered_sales.csv', index=False)
```
